```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt

train_set = pd.read_csv("wine_train.csv")
test_set = pd.read_csv("wine_test.csv")

X_train = np.array(train_set[list(train_set.columns[1:-1])])
y_train = np.array(train_set["quality"])
X_test = np.array(test_set[list(test_set.columns[1:-1])])
y_test = np.array(test_set["quality"])

Test_accuracy =[]
Train_accuracy = []

ridge = Ridge().fit(X_train,y_train)
ridge1 = Ridge(alpha = 1).fit(X_train, y_train)
ridge10 = Ridge(alpha=1000).fit(X_train,y_train)
ridge01 = Ridge(alpha=.1).fit(X_train,y_train)
plt.plot(ridge1.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge10.coef_, '^', label="Ridge alpha=1000")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.1")
plt.plot(ridge.coef_, 'o', label="LinearRegression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.hlines(0, 0, len(ridge.coef_))
plt.ylim(-10, 10)
plt.legend()

print("Test_accuracy{:.2f}"  .format(ridge1.score(X_test,y_test)))
print("Train_accuracy:{:.2f}" .format(ridge1.score(X_train,y_train)))
```
