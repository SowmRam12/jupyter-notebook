```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso



train_set = pd.read_csv("wine_train.csv")
test_set = pd.read_csv("wine_test.csv")

X_train = np.array(train_set[list(train_set.columns[1:-1])])
y_train = np.array(train_set["quality"])
X_test = np.array(test_set[list(test_set.columns[1:-1])])
y_test = np.array(test_set["quality"])

test_accuracy =[]
training_accuracy=[]


alphas = np.logspace(-5, 5, 100)
for a in alphas:
    reg = Lasso()
    reg.set_params(alpha=a)
    reg.fit(X_train, y_train)
    training_accuracy.append(reg.score(X_train,y_train))
    test_accuracy.append(reg.score(X_test,y_test))


#plt.plot(alpha,training_accuracy,label = "training_accuracy")
#plt.plot(alpha,test_accuracy,label = "test_accuracy")
#plt.ylabel('accuracy')
#plt.xlabel('alpha')
#plt.legend()
#plt.show()
#print("Test_accuracy{:.2f}"  .format(reg.score(X_test,y_test)))
#print("Train_accuracy:{:.2f}" .format(reg.score(X_train,y_train)))


plt.figure(figsize=(10, 6))

#plt.subplot(121)
ax = plt.gca()
ax.plot(alphas,test_accuracy,label = "Test accuracy")
ax.plot(alphas,training_accuracy, label = "Training accuracy")
ax.set_xscale("log")
plt.xlabel("alpha")
plt.ylabel("Accuracy")
#plt.axis("tight")
plt.legend(loc="best")
plt.show()
print("Test_accuracy{:.4f}"  .format(reg.score(X_test,y_test)))
print("Train_accuracy:{:.4f}" .format(reg.score(X_train,y_train)))

```
