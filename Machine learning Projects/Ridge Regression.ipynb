```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
import mglearn
from IPython.display import display

train_set = pd.read_csv("wine_train.csv")
test_set = pd.read_csv("wine_test.csv")

X_train = np.array(train_set[list(train_set.columns[1:-1])])
y_train = np.array(train_set["quality"])
X_test = np.array(test_set[list(test_set.columns[1:-1])])
y_test = np.array(test_set["quality"])

test_accuracy =[]
training_accuracy = []

alphas = np.logspace(-6, 6, 100)
for a in alphas:
    reg = Ridge()
    reg.set_params(alpha=a)
    reg.fit(X_train, y_train)
    training_accuracy.append(reg.score(X_train,y_train))
    test_accuracy.append(reg.score(X_test,y_test))

#line = np.linspace(-3,3,100)

#X = np.linspace(alpha, 2*np.pi, 100)
#Y = np.linspace(0, 2*np.pi, 100)
#fig, ax = plt.subplots()
#ax.plot(X, Y)
#plt.plot(alpha,training_accuracy,label = "training_accuracy")
#plt.plot(alpha,test_accuracy,label = "test_accuracy")
#plt.ylabel('accuracy')
#plt.xlabel('alpha')



plt.figure(figsize=(10, 6))

#plt.subplot(121)
ax = plt.gca()
ax.plot(alphas,test_accuracy,label = "Test accuracy")
ax.plot(alphas,training_accuracy, label = "Training accuracy")
ax.set_xscale("log")
plt.xlabel("alpha")
plt.ylabel("Accuracy")
#plt.axis("tight")
plt.legend(loc="best")
plt.show()
print("Test_accuracy{:.3f}"  .format(reg.score(X_test,y_test)))
print("Train_accuracy:{:.3f}" .format(reg.score(X_train,y_train)))



```
